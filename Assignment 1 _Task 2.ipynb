The average temperature and rainfall throughout the year of each country is plotted below, it can be seen that the maximum increase in temperature is in USA and the minimum increase is in India. Australia and France have higher temperatures by 3.5 and 3 degrees respectively.

The average increase in rainfall is different from the temperature data, India received 2.5mm more rainfall between the two time frames. Rainfall in France increase by 1.5mm, in the USA the increase was almost negligible. However, Australia experienced 0.7mm decrease in rainfall.

```python
countries = ['france','USA','India','Australia']
mean_temp_diff = [FRA_temp_diff.mean(), USA_temp_diff.mean(), IND_temp_diff.mean(), AUS_temp_diff.mean()] 
mean_rain_diff = [FRA_rain_diff.mean(), USA_rain_diff.mean(), IND_rain_diff.mean(), AUS_rain_diff.mean()]
fig = plt.figure(figsize=(10,8))

#Average increase in temperature
ax1 = fig.add_subplot(2,1,1)
ax1.scatter(countries, mean_temp_diff, marker='x',color='red')
ax1.set_title(' Average increase in temperature across continents')
ax1.set_xlabel('Country')
ax1.set_ylabel('Temperature difference(Celcius)')
ax1.grid(True,which='both')

#Average increase in rainfall
ax2 = fig.add_subplot(2,1,2)
ax2.scatter(countries, mean_rain_diff, marker='X',color='red')
ax2.set_title(' Average increase in rainfall across continents')
ax2.set_xlabel('Country')
ax2.set_ylabel('Temperature difference(Celcius)')
ax2.grid(True,which='both')

fig.tight_layout(pad=2.0)
```

By looking at both the graphs above we can see that increase in temperature does not necessarily lead to increase in rainfall. Finding the correlation between the two quantities mentioned above we get negative correlation.

```python
np.corrcoef(mean_temp_diff,mean_rain_diff)[0,1]
```
